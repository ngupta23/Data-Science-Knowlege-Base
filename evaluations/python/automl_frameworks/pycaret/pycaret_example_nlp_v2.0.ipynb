# 1. Load Dataset

```python
from pycaret.datasets import get_data
data = get_data('kiva')
```

```python
data.info()
```

# 2. Setup

```python
from pycaret.nlp import *
nlp1 = setup(data, target = 'en', session_id=42, log_experiment=True, log_plots = True, experiment_name='kiva1')
```

# 3. Create Models

```python
models()
```

```python
lda = create_model('lda')
```

```python
nmf = create_model('nmf', num_topics = 6)
```

# 4. Assign Labels

```python
lda_results = assign_model(lda)
lda_results.head()
```

# 5. Analyze Model

```python
plot_model(lda, plot = 'topic_distribution')
```

```python
evaluate_model(lda)
```

```python
plot_model(nmf, plot = 'topic_distribution')
```

```python
evaluate_model(nmf)
```

```python
for i in range(4):
    plot_model(lda, topic_num=f"Topic {i}")
```

```python
for i in range(4):
    plot_model(lda, plot='bigram', topic_num=f"Topic {i}")
```

```python
plot_model(lda, plot='tsne')
```
